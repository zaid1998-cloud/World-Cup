### Insight 1: Top 10 Batsmen by Total Runs

```python
top_batsmen_runs = batting_summary_df.groupby('Batsman_Name')['Runs'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Batsman_Name', data=top_batsmen_runs, palette='viridis')
plt.title('Top 10 Batsmen by Total Runs in World Cup 2023')
plt.xlabel('Total Runs')
plt.ylabel('Batsman Name')
plt.show()
```

### Insight 2: Top 10 Batsmen by Highest Strike Rate (min 50 balls faced)

```python
# Calculate total balls faced for each batsman
batsman_balls = batting_summary_df.groupby('Batsman_Name')['Balls'].sum().reset_index()
batsman_balls.rename(columns={'Balls': 'Total_Balls_Faced'}, inplace=True)

# Filter batsmen who faced at least 50 balls
eligible_batsmen = batsman_balls[batsman_balls['Total_Balls_Faced'] >= 50]['Batsman_Name']

# Calculate average strike rate for eligible batsmen
eligible_batting_summary = batting_summary_df[batting_summary_df['Batsman_Name'].isin(eligible_batsmen)]

top_batsmen_sr = eligible_batting_summary.groupby('Batsman_Name')['Strike_Rate'].mean().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Strike_Rate', y='Batsman_Name', data=top_batsmen_sr, palette='magma')
plt.title('Top 10 Batsmen by Strike Rate (Min. 50 Balls Faced)')
plt.xlabel('Average Strike Rate')
plt.ylabel('Batsman Name')
plt.show()
```

### Insight 3: Top 10 Batsmen by Total Fours and Sixes

```python
top_batsmen_boundaries = batting_summary_df.groupby('Batsman_Name')[['4s', '6s']].sum().reset_index()
top_batsmen_boundaries['Total_Boundaries'] = top_batsmen_boundaries['4s'] + top_batsmen_boundaries['6s']

top_batsmen_boundaries = top_batsmen_boundaries.nlargest(10, 'Total_Boundaries')

plt.figure(figsize=(14, 7))
sns.barplot(x='Total_Boundaries', y='Batsman_Name', data=top_batsmen_boundaries, palette='cividis')
plt.title('Top 10 Batsmen by Total Fours and Sixes')
plt.xlabel('Total Fours and Sixes')
plt.ylabel('Batsman Name')
plt.show()
```

### Insight 4: Top 10 Bowlers by Total Wickets

```python
top_bowlers_wickets = bowling_summary_df.groupby('Bowler_Name')['Wickets'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Wickets', y='Bowler_Name', data=top_bowlers_wickets, palette='viridis')
plt.title('Top 10 Bowlers by Total Wickets in World Cup 2023')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler Name')
plt.show()
```

### Insight 5: Top 10 Bowlers by Best Economy Rate (Min. 10 Overs Bowled)

```python
# Calculate total overs bowled for each bowler
bowler_overs = bowling_summary_df.groupby('Bowler_Name')['Overs'].sum().reset_index()
bowler_overs.rename(columns={'Overs': 'Total_Overs_Bowled'}, inplace=True)

# Filter bowlers who bowled at least 10 overs
eligible_bowlers = bowler_overs[bowler_overs['Total_Overs_Bowled'] >= 10]['Bowler_Name']

# Calculate average economy rate for eligible bowlers
eligible_bowling_summary = bowling_summary_df[bowling_summary_df['Bowler_Name'].isin(eligible_bowlers)]

top_bowlers_economy = eligible_bowling_summary.groupby('Bowler_Name')['Economy'].mean().nsmallest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Economy', y='Bowler_Name', data=top_bowlers_economy, palette='magma')
plt.title('Top 10 Bowlers by Best Economy Rate (Min. 10 Overs Bowled)')
plt.xlabel('Average Economy Rate')
plt.ylabel('Bowler Name')
plt.show()
```

### Insight 6: Top 10 Bowlers by Most Maidens

```python
top_bowlers_maidens = bowling_summary_df.groupby('Bowler_Name')['Maidens'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Maidens', y='Bowler_Name', data=top_bowlers_maidens, palette='cividis')
plt.title('Top 10 Bowlers by Most Maidens in World Cup 2023')
plt.xlabel('Total Maidens')
plt.ylabel('Bowler Name')
plt.show()
```

### Data Cleaning: Handling Missing Values and Data Type Conversions

```python
# Handle missing values in batting_summary_df
# Fill 'Dismissal' NaNs with 'Not Out' as a reasonable assumption for missing dismissal records
batting_summary_df['Dismissal'].fillna('Not Out', inplace=True)

# Fill 'Strike_Rate' NaNs with 0.0, assuming players with missing strike rates didn't face balls or didn't score significantly
batting_summary_df['Strike_Rate'].fillna(0.0, inplace=True)

print('--- Info for batting_summary_df after cleaning missing values ---')
batting_summary_df.info()

print('\n--- Missing values in batting_summary_df after cleaning ---')
display(batting_summary_df.isnull().sum().to_frame(name='Missing Values'))
```

```python
# Handle missing values in world_cup_players_info_df
# 'image_of_player' and 'description' are not critical for numerical analysis, fill with 'N/A'
world_cup_players_info_df['image_of_player'].fillna('N/A', inplace=True)
world_cup_players_info_df['description'].fillna('N/A', inplace=True)

# 'bowlingStyle' is categorical, fill with 'Unknown'
world_cup_players_info_df['bowlingStyle'].fillna('Unknown', inplace=True)

print('--- Info for world_cup_players_info_df after cleaning missing values ---')
world_cup_players_info_df.info()

print('\n--- Missing values in world_cup_players_info_df after cleaning ---')
display(world_cup_players_info_df.isnull().sum().to_frame(name='Missing Values'))
```

```python
# Convert 'Date' column in match_schedule_results_df to datetime objects
# First, standardize the year (assuming current year or 2023 for World Cup)
match_schedule_results_df['Date'] = match_schedule_results_df['Date'].apply(lambda x: x + ' 2023' if 'October' in x or 'November' in x else x)
match_schedule_results_df['Date'] = pd.to_datetime(match_schedule_results_df['Date'], errors='coerce')

print('--- Info for match_schedule_results_df after Date conversion ---')
match_schedule_results_df.info()

print('\n--- Missing values in match_schedule_results_df after Date conversion ---')
display(match_schedule_results_df.isnull().sum().to_frame(name='Missing Values'))
```
